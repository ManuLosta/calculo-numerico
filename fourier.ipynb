```python
import numpy as np
import matplotlib.pyplot as plt
import math

def fourier(m, x):
    sum = 0
    for k in range(1, m, 2):
        sum += 1/k * math.sin(k * x)
    return sum

x = np.linspace(-4*math.pi, 4*math.pi, 400)
m = 1000
y = [fourier(m, xi) for xi in x] 

plt.plot(x, y, label='Función Fourier', color='b')
plt.grid(True)
```

```python
y  = [1.0004, -0.119, 1.5987, 0.2115, -0.6567, -0.3514, -1.6824]
x = [0, 0.897597901, 1.795195802, 2.692793703, 3.590391604, 4.487989505, 5.385587406]
N = 3
    
a_0 = 0
for j in range(2*N + 1):
    a_0 += (1 / (2*N + 1)) * y[j]
a_0 = a_0

a_1 = 0
for j in range(2*N + 1):
    a_1 += (2 / (2*N + 1)) * y[j] * math.cos(x[j])

b_1 = 0
for j in range(2*N + 1):
    b_1 += (2 / (2*N + 1)) * y[j] * math.sin(x[j])

print(a_0)
print(a_1)
print(b_1)
```

```python
f_1 = lambda x: a_0 + a_1*np.cos(x) + b_1*np.sin(x)

x_1 = np.linspace(0, 2*math.pi, 200)
y_1 = f_1(x_1)

plt.plot(x, y, 'o')
plt.plot(x_1, y_1)
```

```python
def fourier(t, m, x, y):
    sum = 0
    N = int((len(x) - 1) / 2)
    
    a_0 = 0
    for j in range(2*N + 1):
        a_0 += (1 / (2*N + 1)) * y[j]
        
    sum += a_0
    
    for k in range(1, m + 1):
        a = 0
        for j in range(2*N + 1):
            a +=  y[j] * math.cos(k * x[j])
            
        
        sum += (2 / (2*N + 1)) * a*math.cos(k*t)
        
        b = 0
        for j in range(2*N + 1):
            b +=  y[j] * math.sin(k * x[j])
            
        sum += (2 / (2*N + 1)) * b *math.sin(k*t)
        
    return sum

y_1 = [fourier(xi, 1, x, y) for xi in x_1]
y_2 = [fourier(xi, 2, x, y) for xi in x_1]
y_3 = [fourier(xi, 3, x, y) for xi in x_1]


plt.plot(x_1, y_1)
plt.plot(x_1, y_2)
plt.plot(x_1, y_3)

plt.plot(x, y, 'o')
```
