```python
import numpy as np
import matplotlib.pyplot as plt
import math

def fourier(m, x):
    sum = 0
    for k in range(1, m, 2):
        sum += 1/k * math.sin(k * x)
    return sum

x = np.linspace(-4*math.pi, 4*math.pi, 400)
m = 1000
y = [fourier(m, xi) for xi in x] 

plt.plot(x, y, label='Función Fourier', color='b')
plt.grid(True)
```
