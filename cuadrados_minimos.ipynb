ln r = ln b + a * ln w

```python
import math

w = [0.017, 0.174, 1.74, 5.45, 0.025, 0.211, 3.02, 4.58, 0.02, 0.171, 3.04, 0.025, 0.783, 1.69, 4.83, 0.087, 1.11, 4.09, 5.96, 0.111, 0.999, 4.28, 4.68, 0.085, 1.29, 4.29, 0.233, 1.35, 2.75, 5.53]

r = [0.154, 0.363, 2.23, 3.52, 0.23, 0.366, 2.01, 2.96, 0.181, 0.334, 3.59, 0.234, 1.47, 1.44, 4.66, 0.296, 0.531, 3.58, 2.4, 0.357, 0.771, 3.28, 5.1, 0.26, 0.87, 3.4, 0.537, 2.48, 1.84, 6.94]
```

Siendo y = ln r y x = ln w

```python
x = [math.log(x) for x in w]
y = [math.log(x) for x in r]

for i in range(len(x)):
    print("x = " + str(x[i]) + " y = " + str(y[i]))
```

```python
print(len(x), len(y))
```

Calculo <x, 1>

```python
sum = 0
for xi in x:
    sum += xi
print(sum)
```

<x, x>

```python
sum = 0
for xi in x:
    sum += xi**2
print(sum)
```

<y, 1>

```python
sum = 0
for yi in y:
    sum += yi
print(sum)
```

<y, x>

```python
sum = 0
for i in range(len(x)):
    sum += x[i] * y[i]
print(sum)
```
